# Exploratory Data Analysis

1] Age Group :

```python
plt.figure(figsize=(4, 4))
sns.histplot(mr['Age'],bins=[15,20,25,30,35],color='hotpink')
plt.title("Heart attack chances by Age",fontsize=15)
plt.show()
```

Conclusion--We can conclude that the chances of heart attck are more in 30-35 age group and we also say that the  as age incrases chances of heart attack increases 

2]Gender :

```python
#cheak for counts based on gender
ax = sns.countplot(x ='Gender',hue='Gender',data=mr,palette='deep')
plt.title("Based on Gender",fontsize=15)
for bars in ax.containers:
    ax.bar_label(bars)
ax.legend(title='Gender', loc='best',labels=['Male','Female','Other'])
plt.show()
```

Conclusion -- By above graph we conclude that the chances of heart attack may be not depends on Gender..

3]Region :

```python
mr['Region'].unique()
```

```python
ax = sns.countplot(x='Region',data=mr,hue='Heart Attack Likelihood',palette='muted')
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Based on Region",fontsize=15)
#ax.legend(title='Gender', loc='best',labels=['East', 'North', 'West', 'North-East', 'Central', 'South'])
plt.show()
```

Conclusion--According to 'Region' we can say the chances of heart attack not depends on region or where you live ..

4] Area :

```python
ax = sns.countplot(x='Urban/Rural',data=mr,hue='Urban/Rural',palette='muted',width=0.5)
plt.title("Based on Area",fontsize=15)
for bars in ax.containers:
    ax.bar_label(bars)
ax.legend(title='Area', loc='best',labels=['Urban','Rural'])
plt.show()
```

Conclusion--According to above graph the chnages of getting heart attack is more in Urban area than Rural Area .It is possible beacause of pollution,noises and internet radiation.

8]Family History :-

```python
ax = sns.countplot(x='Family History of Heart Disease',data=mr,hue='Heart Attack Likelihood',width=0.5)
plt.title("Heart Attack Analysis on Family history of heart disease",fontsize=15)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

Conclusion--Above graph shows that there is still a significant number of individuals with no family history who are at risk of a heart attack and also  the majority of individuals with a family history are unlikely to experience a heart attack so it means the chance of getting heart attack are not depends upon family history of heart disease.

9] Diet Type:-

```python
mr['Diet Type'].unique()
```

```python
ax = sns.countplot(x='Diet Type',data=mr,width=0.5,hue='Heart Attack Likelihood',palette='deep')
plt.title("Heart Attack Analysis by Diet Type",fontsize=15)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

Conclusion--Non-vegetarian diets show the highest count of individuals at risk of heart attacks, suggesting a potential link between meat consumption and heart attack risk.
Vegan diets appear to be associated with the lowest heart attack likelihood, indicating a potential protective effect.
Vegetarian diets show an intermediate pattern, with fewer individuals at risk compared to non-vegetarians but more than vegans.

10]Stress level:-

```python
ax = sns.countplot(x='Stress Level',data=mr,width=0.5,hue='Heart Attack Likelihood')
plt.title("Heart Attack Analysis by Stress level",fontsize=15)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

Conclusion -- Individuals with high stress levels appear to be at a greater proportional risk of heart attacks compared to those with medium or low stress levels.

11]Diabetes :-

```python
ax = sns.countplot(x='Diabetes',data=mr,hue='Heart Attack Likelihood')
plt.title("Heart Attack  by Diabetes",fontsize=15)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

Conslusion -- Among individuals without diabetes, a significantly higher count did not experience a heart attack, compared to a smaller proportion who did.
For individuals with diabetes, the number of those who did not experience a heart attack is still higher than those who did, but the difference is less pronounced compared to non-diabetics.
