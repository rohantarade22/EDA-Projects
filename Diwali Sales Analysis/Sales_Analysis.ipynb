```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding='windows-1252')
df
```

```python
df.isnull().sum()
```

```python
#drop numm rows 
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#drop above remaining null values 
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
#change the data type of float to int
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
#cheak which gender doing more shopping
ax = sns.countplot(x="Gender",data=df,width=0.5,palette={'F':'orange','M':'yellow'},hue='Gender')
plt.title("count by Gender",style='oblique')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#cheak for which gender spend more amount 
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen,palette={'F':'r','M':'orange'},hue="Gender",saturation=0.5,width=0.2)
plt.title("Amount spending by gender")
plt.legend(labels=["Female", "Male"])
plt.show()
```

```python
#Maximum buyers comes under age group
ax = sns.countplot(x="Age Group",data=df,hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Age")
plt.show()
```

```python
#ckeaking for which age group spend more 
sales_age = df.groupby(['Age Group'],as_index=False,observed=True)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age,hue="Age Group",saturation=0.5,width=0.5)
plt.title("Amount spending on age group")
plt.show()
```

```python
#order by states --
order_state = df.groupby(['State'],as_index=False,observed=True)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)
sns.catplot(x='State',y='Orders',data=order_state,hue="State",saturation=0.5,width=0.5,kind='bar')
plt.title("oders by state")

plt.show()
```

```python
#spend amount by state 
order_state = df.groupby(['State'],as_index=False,observed=True)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.catplot(x='State',y='Amount',data=order_state,hue="State",saturation=0.5,width=0.5,kind='bar')
plt.title("Amount by state")
plt.xticks(rotation=90)
plt.show()
```

```python
#marrital status and order
ax = sns.countplot(x="Marital_Status",data=df,width=0.2,palette=['orange','yellow'],hue="Marital_Status")
ax.set_xticks([0, 1])
ax.set_xticklabels(['married', 'unmarried'])
plt.legend(labels=['married', 'unmarried'])
plt.show()
```

```python
#marrital status and amount
order_state = df.groupby(['Marital_Status','Gender'],as_index=False,observed=True)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Marital_Status",y="Amount",data=df,hue="Gender",width=0.2)
plt.show()
```
